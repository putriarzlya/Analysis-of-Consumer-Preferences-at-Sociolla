# DATA CLEANING

- Pada tahap data cleaning ini yang dilakukan pertama yaitu membuat dataframe baru (copy) kemudian split kolom brand_name menjadi brand_id dan brand_name.
- Memisahkan(split) kolom price_range karena beberapa nilainya terdapat yang bentuk range.
- Split kolom "categories" menjadi category, type_of_categories, product_of_categories karena secara keseluruhan merupakan hal yang sama maka dari itu nantinya hanya diambil kolom "category".

```python
# Membuat dataframe baru dengan meng-copy dataframe sebelumnya
df_2 = df.copy()

# Memisahkan brand_id dan brand_name
df_2[['brand_id', 'brand_name']] = df_2['brand_name'].str.split('_', n=1, expand=True)
# # Memisahkan range price_range
df_2[['price_range', 'price_range2']] = df_2['price_range'].str.split('-', n=1, expand=True)
# Memisahkan antara kategori dan produk di kolom categories
df_2[['category', 'type_of_categories', 'product_of_categories']] = df_2['categories'].str.split(';', n=2, expand=True)

df_2
```

- Menghilangkan "Rp" dan "." pada kolom harga agar dapat dilakukan perhitungan statistik.
- Menghilangkan "$" pada kolom product name karena karakter khusus tersebut tidak bisa dimunculkan di plot.

```python
# Menghilangkan "Rp" dan "." pada kolom price_range & price_range2

df_2['price_range'] = pd.to_numeric(df_2['price_range'].str.replace('Rp', '', regex=False).str.replace('[^\d]', '', regex=True))
df_2['price_range2'] = pd.to_numeric(df_2['price_range2'].str.replace('Rp', '', regex=False).str.replace('[^\d]', '', regex=True))

# Membersihkan nama produk dari karakter khusus seperti "$"
df_2['product_name'] = df_2['product_name'].str.replace('$', 's')

df_2.head()
```

Setelah menghilangkan "Rp" dan karakter khusus pada kolom kedua kolom price_range dan kolom product_name, kemudian selanjutnya mencari nilai median dari kedua kolom tersebut karena beberapa produk memiliki range harga yang kemudian dimasukkan ke kolom baru bernama "price"

# Analisis & Perhitungan

### Kolom Price
Menghitung rata-rata, median, standar deviasi, skewness, dan kurtosis

```python
# Mean, median, standar deviasi dari kolom "price" menggunakan method .describe()
desc = df_2['price'].describe()
print (desc)

# Menghitung skewness menggunakan method .skew() dari kolom "price"
skewness = df_2['price'].skew()
print(f'Nilai skewness yang diperoleh dari kolom harga produk yaitu {skewness}')

# Menghitung kurtosis menggunakan method .kurtosis() dari kolom "price"
kurtosis_value = df_2['price'].kurtosis()
print(f'Nilai kurtosis yang diperoleh dari kolom harga produk yaitu {kurtosis_value}')

print("--------------------------------------------------------------------------------")

# Mengetahui adanya outlier atau tidak dengan teknik Tukey's Rule karena nilai skewness yang diperoleh dari kolom price lebih besar dari 0.5
q1 = df_2.price.quantile(0.25) # mencari value Q1
q3 = df_2.price.quantile(0.75) # mencari value Q3

iqr = q3 - q1 # mencari value interquartile range
min_boundary = q1 - (1.5 * iqr) # mencari value lower boundary
max_boundary = q3 + (1.5 * iqr) # mencari value upper boundary

print(f'value iqr adalah: {iqr}')
print(f'minimum value adalah: {min_boundary}')
print(f'maximum value adalah: {max_boundary}')
print("--------------------------------------------------------------------------------")

# Mencari outlier dan data non-outlier
outliers = df_2.price[(df_2.price < min_boundary) | (df_2.price > max_boundary)]
non_outliers = df_2.price[(df_2.price >= min_boundary) & (df_2.price <= max_boundary)]

# Menghitung persentase outlier
percentage_outliers = len(outliers) * 100 / len(df_2.price)

# Mencetak persentase outlier dengan format dua angka di belakang koma
print(f'Persentase outliers: {percentage_outliers:.2f}%')
print("--------------------------------------------------------------------------------")

# Sebelum outlier
mean = df_2.price.mean() # menghitung mean/rata-rata
median = df_2.price.median() # menghitung median
mode = df_2.price.mode()[0] # menghitung modus

print(f'mean sebelum filtering data adalah: ', mean)
print(f'median sebelum filtering data adalah:  ', median)
print(f'mode sebelum filtering data adalah:  ', mode)
print("--------------------------------------------------------------------------------")

# Setelah penghapusan outlier
no_mean = non_outliers.mean() # mean / average / rata-rata
no_median = non_outliers.median() # median / nilai tengah
no_mode = non_outliers.mode()[0] # modus / paling sering muncul

print(f'mean setelah outlier filtering data adalah: ', no_mean)
print(f'median setelah outlier filtering data adalah: ', no_median)
print(f'mode setelah outlier filtering data adalah: ', no_mode)
print("--------------------------------------------------------------------------------")
```

- Nilai standar deviasi yaitu 2.21 yang dimana nilainya lebih besar sedikit dari nilai mean. Hal ini mengindikasikan bahwa data tersebar cukup luas disekitar rata-ratanya.
- Diperoleh nilai minimum 105.00 dan nilai maksimal 3,077,000 yang mana kedua nilai tersebut sangat jauh berbeda.

ada kecenderungan yang jelas, misalnya, semakin tinggi rating rata-rata produk, semakin tinggi juga jumlah pelanggan yang berniat untuk melakukan pembelian ulang, maka Anda bisa menyimpulkan bahwa kualitas produk berkorelasi positif dengan keinginan pelanggan untuk melakukan pembelian ulang.

dilakukan uji korelasi kendall

```python
# Using Pandas

fig, ax = plt.subplots()

ax.set_title('Pengaruh Penilaian Produk Terhadap Keputusan Pembelian')
ax.set_xlabel('Rating')
ax.set_ylabel('Jumlah Repurchase')

# Plot scatter plot
ax.scatter(df_2['average_rating'], df_2['total_repurchase_yes_count'], label='Data Produk', color='blue', alpha=0.7)

# Menambahkan grid
ax.grid(True)

# Menampilkan legenda
ax.legend()

plt.show()
```

**Insight:**

1. Terdapat pola yang menunjukan bahwa produk dengan rating tinggi (4-5) cenderung memiliki jumlah pembelian yang tinggi. Hal ini dapat dikatakan kepuasan pelanggan berkontribusi pada loyalitas dan repurchase.

2. Menunjukan korelasi positif sehingga pentingnya meningkatkan kualitas produk untuk meningkatkan tingkat kepuasan pelanggan dan dapat memperkuat loyalitas mereka. Perusahaan dapat fokus pada pembaharuan produk, layanan pelanggan, dan aspek lainnya yang berpengaruh.

3. Terkait segmentasi pelanggan, dapat dikatakan bahwa pelanggan yang cenderung memberikan rating tinggi dan melakukan repurchase dapat dianggap sebagai pelanggan setia yang layak mendapat perhatian khusus dalam strategi pemasaran dan retensi pelanggan.

4. Terdapat produk dengan rating tinggi tetapi jumlah repurchase rendah. Hal ini menunjukan bahwa terdapat faktor lain diluar kualitas produk yang mempengaruhi keputusan pembeli.

Agar korelasi ini lebih valid, maka dilanjutkan dengan uji Kendall Tau.

```python
# Uji Kendall
korelasi_koefisien, p_value = kendalltau(df['total_repurchase_yes_count'], df['average_rating'])
print(f'Nilai koefisien Kendall Tau: {korelasi_koefisien}')
print(f'P-value: {p_value}')

if p_value < 0.05:
    print('Korelasi signifikan. H0 berhasil ditolak.')
else:
    print('Tidak ada korelasi signifikan. H0 gagal ditolak.')
```

Berdasarkan uji Kendall Tau, korelasi antara rating dengan jumlah repurchase memiliki korelasi signifikan.

## Uji Inferensial

Dalam kasus ini digunakannya uji hipotesis, yaitu metode statistik yang digunakan untuk membuat keputusan tentang sifat-sifat suatu populasi berdasarkan data sampel. 
- H0 = tidak ada korelasi signifikan antara rating dengan ulasan

- H1 = terdapat korelasi signifikan antara rating dengan ulasan

- Jenis hipotesis yang digunakan dalam kasus ini yaitu melihat adanya/tidak adanya korelasi antara rating dengan ulasan maka menggunakan uji hipotesis dua sampel independen.

```python
# Using Seaborn
fig, ax = plt.subplots()

ax.set_title('Pengaruh Ulasan Terhadap Penilaian Produk')
ax.set_xlabel('Total Review')
ax.set_ylabel('Rating')

sns.scatterplot(x=df_2['total_reviews'],y=df_2['average_rating'],label='Data Produk', color='blue', alpha=0.7, ax=ax)

# Menambahkan grid
ax.grid(True)

# Menampilkan legenda
ax.legend()

plt.show()

stat, p_value = wilcoxon(df_2['total_reviews'], df_2['average_rating'])
# Menampilkan hasil
print('Statistics=%.3f' % stat)
print('p_value=%.3f' % p_value)

# Menginterpretasikan hasil
if p_value < 0.05:
    print('Korelasi signifikan. H0 berhasil ditolak.')
else:
    print('Tidak ada korelasi signifikan. H0 gagal ditolak.')
```

**Insight:**
1. Terdapat pola yang dapat diamati yaitu produk yang memiliki lebih banyak ulasan, nilai ratingnya juga tinggi. Ini bisa menunjukkan bahwa produk dengan kualitas yang baik cenderung menerima lebih banyak ulasan.

2. Pentingnya mendengarkan umpan balik pelanggan karena total ulasan yang tinggi dapat menjadi indikator kepuasan pelanggan dan kepercayaan konsumen terhadap produk.

3. Meskipun ada korelasi antara kedua hal tersebut, masih ada variasi dalam penilaian pelanggan. Beberapa produk mungkin memiliki rating yang tinggi meskipun jumlah ulasannya tidak sebanyak yang lain, dan sebaliknya.

4. Berdasarkan data, sejauh ini produk yang mendapatkan rating rendah memiliki sedikit ulasan. Dapat dikatakan bahwa belum ada produk yang perlu pengawasan (dalam kata lain, jika kualitas produk tersebut rendah dan mendapatkan ulasan banyak).

5. Data ini juga dapat memberikan wawasan tentang bagaimana ulasan produk mempengaruhi keputusan pembelian pelanggan. Produk dengan rating dan ulasan yang tinggi cenderung lebih menarik bagi calon pembeli.

## Uji Diferensial

Uji yang dipilih untuk menentukan adanya korelasi/tidak yaitu uji Kendall Tau karena data yang terdistribusi tidak secara normal dan juga bentuk data tidak perlu continous dan diskret tetapi harus salah satunya ordinal.

```python
# Using Seaborn
fig, ax = plt.subplots()

ax.set_title('Pengaruh Ulasan Terhadap Produk di Wishlist')
ax.set_xlabel('Total Review')
ax.set_ylabel('Total Wishlist')

sns.scatterplot(x=df_2['total_reviews'],y=df_2['total_in_wishlist'],label='Data Produk', color='blue', alpha=0.7, ax=ax)

# Menambahkan grid
ax.grid(True)

# Menampilkan legenda
ax.legend()

plt.show()

# Kelompokkan data berdasarkan produk dan hitung jumlah ulasan serta jumlah produk yang dimasukkan ke dalam daftar keinginan
product_grouped = df_2.groupby('product_name').agg({'total_reviews': 'mean', 'total_in_wishlist': 'mean'})

# Hitung korelasi Kendall
kendall_corr, p_value = kendalltau(product_grouped['total_reviews'], product_grouped['total_in_wishlist'])

print("Korelasi Kendall antara jumlah ulasan dan jumlah produk di daftar keinginan:", kendall_corr)
print("Nilai p:", p_value)

if p_value < 0.05:
    print('Korelasi signifikan. H0 berhasil ditolak.')
else:
    print('Tidak ada korelasi signifikan. H0 gagal ditolak.')
```

**Insight:**
1. Produk yang memiliki sedikit ulasan dan jumlah produk dalam daftar keinginan rendah mungkin merupakan produk baru atau kurang dikenal di platform tersebut. Hal ini menunjukkan bahwa pelanggan mungkin belum banyak memperhatikan produk-produk tersebut.

2. Produk dengan jumlah ulasan rendah namun jumlah produk dalam daftar keinginan yang tinggi mungkin menunjukkan adanya peluang pertumbuhan. Meskipun produk tersebut belum banyak direview, namun minat pelanggan terhadapnya cukup tinggi, yang bisa dijadikan sebagai sasaran untuk promosi lebih lanjut.

3. Produk dengan jumlah ulasan yang tinggi dan jumlah produk dalam daftar keinginan yang rendah mungkin menunjukkan bahwa promosi dan pengiklanan produk tersebut belum efektif dalam menarik minat pelanggan. Diperlukan strategi promosi yang lebih kuat untuk meningkatkan kesadaran dan minat pelanggan terhadap produk tersebut
